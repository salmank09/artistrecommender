```python
def load_artist_followers():
    artist_follower_writer = csv.writer(open('artist_follower.csv','wb'))
    followers_writer = csv.writer(open('followers.csv','wb'))

    artist_follower_writer.writerow(['exception', 'artist_num', 'artist_follower_num', 'artist_alias','follower_alias'])
    followers_writer.writerow(['exception', 'artist_num', 'artist_follower_num','artist_alias', 'follower_alias', 'comments_count', 'followers_count', 'followings_count', 'last_modified', 'likes_count', 'plan', 'playlist_count', 'public_favorites_count', 'reposts_count', 'track_count', 'uri', 'username'])

    artist_num = 0
    writer = csv.writer(open('artist_follower.csv','a'))

    for artist in artist_followers.find():
        artist_count = artist_followers.count()
        artist_follower_num = 0
        artist_num += 1
        followers_count = artist['followers_count']
        for follower in artist["followers"]:
            artist_follower_num += 1
            if artist_follower_num % 5000 == 0:
                print "{} of {} followers complete for alias {}, artist {} of {}".format(artist_follower_num, followers_count, artist['alias'], artist_num, artist_count)                
            try:
                artist_follower_writer.writerow(['no exception', str(artist_num), str(artist_follower_num), artist["alias"].encode("utf-8"), follower["permalink"].encode("utf-8")])
                followers_writer.writerow(['no exception', str(artist_num), str(artist_follower_num), artist["alias"].encode("utf-8"), follower["permalink"].encode("utf-8"), follower['comments_count'], follower['followers_count'], follower['followings_count'], follower['last_modified'], follower['likes_count'], follower['plan'], follower['playlist_count'], follower['public_favorites_count'], follower['reposts_count'], follower['track_count'], follower['uri'], follower['username'].encode("utf-8")])
            except Exception as e:
                print str(e), [str(artist_num), str(artist_follower_num), artist["alias"].encode("utf-8"), follower["permalink"].encode("utf-8"), follower['comments_count'], follower['followers_count'], follower['followings_count'], follower['last_modified'], follower['likes_count'], follower['plan'], follower['playlist_count'], follower['public_favorites_count'], follower['reposts_count'], follower['track_count'], follower['uri'], follower['username']]
                artist_follower_writer.writerow([str(e), str(artist_num), str(artist_follower_num), artist["alias"].encode("utf-8"), follower["permalink"].encode("utf-8")])
                followers_writer.writerow([str(e), str(artist_num), str(artist_follower_num), artist["alias"].encode("utf-8"), follower["permalink"].encode("utf-8")])
                continue
        print "{} of {} followers complete for alias {}, artist {} of {}".format(artist_follower_num, followers_count, artist['alias'], artist_num, artist_count)
```
